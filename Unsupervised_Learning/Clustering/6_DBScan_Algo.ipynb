```python
from sklearn.preprocessing import StandardScaler,normalize
scaled = StandardScaler()
scaled_data = scaled.fit_transform(df)
normalize_data = normalize(scaled_data)
```

```python
df1 = pd.DataFrame(normalize_data,columns=['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES',
       'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PURCHASES_FREQUENCY',
       'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY',
       'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX',
       'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT',
       'TENURE'])
```

```python
df1.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_principal = pca.fit_transform(df1)
```

```python
df_final = pd.DataFrame(x_principal,columns=['P1','P2'])
```

```python
df_final.head()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.0375,min_samples=3).fit(df_final)  # you acn take eps value as per your decision also min samples.
labels = dbscan.labels_
np.unique(labels)
```

```python
r = plt.scatter(df_final['P1'],df_final['P2'],color = 'r')
g = plt.scatter(df_final['P1'],df_final['P2'],color = 'g')
b = plt.scatter(df_final['P1'],df_final['P2'],color = 'b')
k = plt.scatter(df_final['P1'],df_final['P2'],color = 'k')
plt.show
```

```python
colours = {}
colours[0] = 'r'
colours[1] = 'g'
colours[2] = 'b'
colours[-1] = 'k'
cvec = [colours[label] for label in labels]
plt.figure(figsize=(9,9))
plt.scatter(df_final['P1'],df_final['P2'],c=cvec)
plt.legend((r,g,b,k),('Label 0','Label 1','Label -1','Label 2'))
plt.show()
```
