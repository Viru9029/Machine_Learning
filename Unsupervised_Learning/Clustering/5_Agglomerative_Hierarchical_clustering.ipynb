```python
#Perform StandardScalar to scaled data and Perform the Normalize for data normalization to plot the dendrogram to know the n_clusters value
from sklearn.preprocessing import StandardScaler,normalize
scaled = StandardScaler()
df_scaled = scaled.fit_transform(df)
df_normalized = normalize(df_scaled)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_normalized)
```

```python
df_pca = pd.DataFrame(df_pca,columns=['PC1','PC2'])
```

```python
df_pca.head()
```

```python
import scipy.cluster.hierarchy as hc
plt.figure(figsize=(10,8))
plt.title('Visualising the data')
dendrogram_plot = hc.dendrogram(hc.linkage(df_pca,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=2)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df_pca['PC1'],df_pca['PC2'],c = agg.fit_predict(df_pca),cmap='rainbow')
plt.title('Agglomerative Hierarchical_clustering - Scatter plot',fontsize = 18)
plt.show()
```
