```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

```python
boston_dataset.keys()
```

```python
boston_dataset.data.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaled = StandardScaler()
scaled_data = scaled.fit_transform(boston_dataset.data)
scaled_data
scaled_target = scaled.fit_transform(boston_dataset.target.reshape(-1, 1))
scaled_target
```

```python
x = scaled_data
x.shape
```

```python
y = scaled_target
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=12)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_regression = DecisionTreeRegressor()
dt_regression.fit(x_train,y_train)
```

```python
from sklearn import tree
fig = plt.figure(figsize=(15,12))
x=tree.plot_tree(dt_regression)
```

```python
y_pred = dt_regression.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error,r2_score
MSE = mean_squared_error(y_test,y_pred)
print('mean square error : ',MSE)

RMSE = np.sqrt(MSE)
print('Root Mean Square Error : ',RMSE)

r2_error = r2_score(y_test,y_pred)
print('r2_square error ',r2_error)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Value')
plt.ylabel('Predicated Value')
plt.title('True value vs predicted value : Decision Tree Algorithm (Regression)')
plt.show
```
